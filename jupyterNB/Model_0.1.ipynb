```python
stand['label'] = 0
```

```python
small_stand = stand[400:1000]
small_stand
```

```python
squat['label'] = 1
```

```python
small_squat = squat[0:600]
small_squat
```

```python
df = pd.concat([small_stand, small_squat], ignore_index=True)
df
```

```python
X = df.drop('label', axis=1)
y = df['label']
```

```python
from sklearn.svm import SVC

clf = SVC(kernel='rbf', gamma=0.001, random_state=42).fit(X, y)
```

```python

pred = clf.predict(X)
tn, fp, fn, tp = confusion_matrix(y, pred).ravel()
tn, fp, fn, tp
```

```python
from micromlgen import port
c_code = port(clf)
#print(c_code)
```

```python
f = open( 'model.txt', 'w' )
f.write(c_code)
f.close()
```

```python
from sklearn.ensemble import RandomForestClassifier
estimator = RandomForestClassifier()
estimator.fit(X, y)
```

```python
pred = estimator.predict(X)
tn, fp, fn, tp = confusion_matrix(y, pred).ravel()
tn, fp, fn, tp
```
